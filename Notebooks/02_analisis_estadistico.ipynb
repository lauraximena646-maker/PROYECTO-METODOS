# 2. Análisis estadístico y visualización

**Objetivo:** describir el comportamiento de temperatura y humedad a partir de los datos limpios y visualizar la evolución temporal de los tres sensores de temperatura.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

ruta_datos = Path('datos_limpios.csv')
df = pd.read_csv(ruta_datos)
df.head()
```

## Estadísticas descriptivas

La tabla resume los valores mínimos, máximos, medias y desviaciones estándar de las variables principales.

```python
variables = ['T_guadua', 'T_plastico', 'T_ambiente', 'H_guadua', 'H_plastico', 'H_ambiente']
stats = df[variables].describe().round(2)
stats
```

## Evolución temporal

Se comparan las temperaturas de la casa de guadua, la casa de plástico y el ambiente exterior a lo largo del tiempo.

```python
plt.figure(figsize=(12, 6))
plt.plot(df['tiempo_min'], df['T_guadua'], label='Casa Guadua', alpha=0.8)
plt.plot(df['tiempo_min'], df['T_plastico'], label='Casa Plástico', alpha=0.8)
plt.plot(df['tiempo_min'], df['T_ambiente'], label='Ambiente', alpha=0.8)
plt.xlabel('Tiempo (minutos)')
plt.ylabel('Temperatura (°C)')
plt.title('Evolución de la temperatura en los tres entornos')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
